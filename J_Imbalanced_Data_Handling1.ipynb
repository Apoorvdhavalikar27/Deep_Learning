## HANDLING IMBALANCED DATA

```python
import pandas as pd
df = pd.read_csv("DataSets/Customer_Churn_Cleaned.csv")
df.drop(columns = ['Unnamed: 0'],inplace = True)
```

```python
df.head()
```

```python
df.Churn.value_counts()
# here class 1 has very low data compared to class 0 ,So it is clearly imbalanced DataSet
```

```python
X = df.drop(columns = ['Churn'])
y = df.Churn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 10)
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
```

```python
def MyNeuralNetwork(X_train,X_test,y_train,y_test,epoch):
    
    model = keras.Sequential([
            keras.layers.Dense(27,input_shape = (27,),activation = 'relu'),
            keras.layers.Dense(18,activation = 'relu'),
            keras.layers.Dense(9,activation = 'relu'),
            keras.layers.Dense(1,activation = 'sigmoid')
    ])
    model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
    
    print(model.fit(X_train,y_train,epochs = epoch))
    
    print(model.evaluate(X_test,y_test))
    
    y_p = model.predict(X_test)
    y_p = np.round(y_p)
    
    truth = y_test
    
    cm = confusion_matrix(truth,y_p)
    print(classification_report(truth,y_p)) 
# Here we define same neural network in function
```

```python
MyNeuralNetwork(X_train,X_test,y_train,y_test,100)
# At the bottom we print classification report
# in which f1-score for class 1 is very low so it is clearly imbalanced data
```

## METHOD 1 :  By Down sampling the dominant Class 

```python
# first we divide two classes in different variables

class_0_count,class_1_count = df.Churn.value_counts()
print(class_0_count,class_1_count)

#Divide classes data also

class_0 = df[df['Churn'] == 0]
class_1 = df[df['Churn'] == 1]
```

```python
# We use sample(i) method , which select random i samples from given dataset
class_0.sample(2)
# Here we select random 2 samples from class 0
```

```python
class_0_downS = class_0.sample(class_1_count)
# Here we down sample class 0 to 1869 values
```

```python
df_test1 = pd.concat([class_0_downS,class_1])
# Here we create first set of data with same(1869) values in both classes
df_test1.Churn.value_counts()
```

```python
X = df_test1.drop(columns = ['Churn'])
y = df_test1.Churn
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 10,stratify = y)
```

```python
y_train.value_counts()
# So extra parameter stratify used to balance the data even in y_test,y_train
```

```python
MyNeuralNetwork(X_train,X_test,y_train,y_test,100)
# Now our f1 score for class 1 is improved
```

## METHOD 2 : By Over sampling the weak Class 

```python
print(class_0_count,class_1_count)
```

```python
# class_1.sample(2000)
class_1_overS = class_1.sample(class_0_count,replace = True)
# We can add random values from that class to over sample that dataset...But replace = True os required for over sampling
class_1_overS.shape
```

```python
df_test2 = pd.concat([class_1_overS,class_0])
df_test2.shape
```

```python
X = df_test2.drop(columns = ['Churn'])
y = df_test2.Churn
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 10,stratify = y)
```

```python
MyNeuralNetwork(X_train,X_test,y_train,y_test,100)
# f1-score is improved for both classes
```

## Method 3 : SMOT (Synthetic Minority Over-sampling Technique)

```python
# pip install imblearn
```

```python
X = df.drop(columns = ['Churn'])
y = df.Churn
```

```python
from imblearn.over_sampling import SMOTE
# SMOTE uses k-means clusterin gto select the samples insted of random selection
```

```python
smote = SMOTE(sampling_strategy = 'minority')
```

```python
X_sm,y_sm = smote.fit_resample(X,y)
```

```python
# y_sm.shape
X_sm.shape
```

```python
y_sm.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_sm,y_sm,test_size = 0.2,random_state = 10)
```

```python
MyNeuralNetwork(X_train,X_test,y_train,y_test,100)
# SMOT achive highest accuracy and f1-score than other two mwthods
```
