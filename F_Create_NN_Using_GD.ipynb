## Create Own Neural Net Using Gradient Descent

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras

df = pd.read_csv('DataSets/DL6_DATA.csv')
df.head()
```

```python
df.shape
```

```python
X = df.drop(columns = ['bought_insurance'])
y = df.bought_insurance
```

```python
# Scaling
X['age'] = X.age/100     
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 10)
```

```python
model = keras.Sequential([
        keras.layers.Dense(1 , input_shape = (2,), activation = 'sigmoid',kernel_initializer = 'ones',bias_initializer = 'zeros')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',                   #log_loss 
    metrics = ['accuracy']

)
```

```python
model.fit(X_train,y_train,epochs = 5000)
```

```python
model.evaluate(X_test,y_test)
```

```python
def sigmoid(WS):
    return 1/(1+np.exp(-WS))

def log_loss(y_predicted,y_true):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
class My_Neural_Network:
    
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0
        
    def fit(self,a_train,b_train,epochs,loss_threshold):
        self.w1,self.w2,self.bias = self.gradient_descent(a_train['age'],a_train['affordibility'],b_train,epochs,loss_threshold)
    
    def predict(self,a_test):
        weighted_sum = self.w1*a_test['age'] + self.w2*a_test['affordibility'] + self.bias
        return sigmoid(weighted_sum)
    
    
    def gradient_descent(self,age,affordibility,y_true,epochs,loss_threshold):
        learning_rate = 0.05
        n = len(age)
        w1 = w2 = 1
        bias = 0

        for i in range(epochs):
            weighted_sum = w1*age + w2*affordibility + bias
            y_predicted = sigmoid(weighted_sum)
            loss = log_loss(y_predicted,y_true)

            dw1 = (1/n)*(np.dot(np.transpose(age),(y_predicted - y_true)))
            dw2 = (1/n)*(np.dot(np.transpose(affordibility),(y_predicted - y_true)))
            dbias = np.mean(y_predicted - y_true)

            w1 = w1 - learning_rate * dw1
            w2 = w2 - learning_rate * dw2
            bias = bias - learning_rate * dbias
            
            if i%250 == 0:
                print(f'epoch:{i},w1:{w1},w2:{w2},bias:{bias},loss:{loss}')

            if loss <= loss_threshold:
                break

        return w1,w2,bias       
```

```python
my_model = My_Neural_Network()
```

```python
my_model.fit(X_train,y_train,5000,0.4123)
```

```python
def gradient_descent(age,affordibility,y_true,epochs,loss_threshold):
    learning_rate = 0.05
    n = len(age)
    w1 = w2 = 1
    bias = 0
    
    for i in range(epochs):
        weighted_sum = w1*age + w2*affordibility + bias
        y_predicted = sigmoid(weighted_sum)
        loss = log_loss(y_predicted,y_true)
        
        dw1 = (1/n)*(np.dot(np.transpose(age),(y_predicted - y_true)))
        dw2 = (1/n)*(np.dot(np.transpose(affordibility),(y_predicted - y_true)))
        dbias = np.mean(y_predicted - y_true)
        
        w1 = w1 - learning_rate * dw1
        w2 = w2 - learning_rate * dw2
        bias = bias - learning_rate * dbias
        
        print(f'epoch:{i},w1:{w1},w2:{w2},bias:{bias},loss:{loss}')
        
        if loss <= loss_threshold:
            break
            
    return w1,w2,bias      
```
